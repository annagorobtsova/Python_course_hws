# Задание 3

Посчитайте процент выживаемости у каждого класса пассажиров (Pclass) 

```python
df["Survived"] = df["Survived"].replace([0,1], ["not survived", "survived"]) # переименовать значения для удобства восприятия (0 - not sirvived; 1 - survived)
```

```python
survival_count = df.groupby("Pclass")["Survived"].value_counts() # кол-во пассажиров в трех классах, которые выжили и не выжили
survival_count
```

```python
survival_prcnt = survival_count.groupby(level=0).apply(lambda x:100 * x / float(x.sum())) # процент пассажиров в трех классах, которые выжили и не выжили (наибольшая доля погибших в 3 классе - 75%, а наименьшая в 1 классе - 37%)
survival_prcnt
```
