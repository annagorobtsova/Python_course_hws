# Работа с API

## Получить информацию обо всех книгах и сохранить в pd.DataFrame - 1 балл

```python
import requests
import pandas as pd
```

```python
api_base_url = 'https://www.anapioficeandfire.com/api'
```

```python
books_list = requests.get(
    f'{api_base_url}/books').json()
```

```python
# Преобразование в Pandas DataFrame
df_books = pd.DataFrame(books_list)
df_books
```

## Получить информацию обо всех домах Вестероса (Houses) и сохранить в pd.DataFrame - 1 балл

```python
houses_list = requests.get(
    f'{api_base_url}/houses').json()
```

```python
# Преобразование в Pandas DataFrame
df_houses = pd.DataFrame(houses_list)
df_houses
```

## Получить информацию обо всех домах Вестероса, у которых есть девиз, и сохранить в pd.DataFrame - 2 балла

```python
parameter = {'hasWords': 'true'}  # Параметр для фильтрации домов с девизами
```

```python
houses_with_words_list = requests.get(
    f'{api_base_url}/houses', params=parameter).json()
```

```python
# Преобразование в Pandas DataFrame
df_houses_with_words = pd.DataFrame(houses_with_words_list)
df_houses_with_words
```
